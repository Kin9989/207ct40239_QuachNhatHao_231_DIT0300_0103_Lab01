```python
# câu 4
import cv2 as cv
import matplotlib.pyplot as hao



image_path = "/content/Screenshot 2023-11-05 184843.png"
image = cv.imread(image_path)

# Phóng to ảnh (ví dụ: phóng to gấp đôi)
scale_factor = 2
height, width, channels = image.shape
new_height = int(height * scale_factor)
new_width = int(width * scale_factor)

# Hiển thị ảnh phóng to
hao.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB), extent=[0, new_width, 0, new_height])
hao.title('Ảnh sau khi phóng to')
hao.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as hao

image_path = "/content/Screenshot 2023-11-05 184843.png"
image = cv.imread(image_path)

# Xác định vùng muốn crop (ví dụ: crop một phần giữa ảnh)
start_row, end_row = 100, 300  # Phạm vi dòng (row)
start_col, end_col = 200, 400  # Phạm vi cột (column)

# Crop ảnh
cropped_image = image[start_row:end_row, start_col:end_col]

# Hiển thị ảnh gốc
hao.subplot(1, 2, 1)
hao.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
hao.title('Ảnh gốc')

# Hiển thị ảnh đã crop
hao.subplot(1, 2, 2)
hao.imshow(cv.cvtColor(cropped_image, cv.COLOR_BGR2RGB))
hao.title('Ảnh đã crop')

hao.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as hao
import numpy as np

image1_path = "/content/Screenshot 2023-11-05 184843.png"
image2_path = "/content/saved_image.jpg"

image1 = cv.imread(image1_path)
image2 = cv.imread(image2_path)

# Kiểm tra kích thước của ảnh
height1, width1, channels1 = image1.shape
height2, width2, channels2 = image2.shape

# Ghép ngang
concatenated_horizontal = np.concatenate((image1, image2), axis=1)

# Ghép dọc
concatenated_vertical = np.concatenate((image1, image2), axis=0)

# Hiển thị ảnh gốc
hao.subplot(1, 3, 1)
hao.imshow(cv.cvtColor(image1, cv.COLOR_BGR2RGB))
hao.title('Ảnh 1')
hao.subplot(1, 3, 2)
hao.imshow(cv.cvtColor(image2, cv.COLOR_BGR2RGB))
hao.title('Ảnh 2')
# Hiển thị ảnh ghép ngang
hao.subplot(1, 3, 3)
hao.imshow(cv.cvtColor(concatenated_horizontal, cv.COLOR_BGR2RGB))
hao.title('Ghép ngang')

hao.show()

# Hiển thị ảnh ghép dọc
hao.subplot(1, 2, 1)
hao.imshow(cv.cvtColor(concatenated_vertical, cv.COLOR_BGR2RGB))
hao.title('Ghép dọc')
hao.show()
```
